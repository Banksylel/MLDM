## Mushroom feature distribution plots
The following plots were generated for the purposes of data exploration.

#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load data

```python
mushroom_data = pd.read_csv('mushrooms.csv')
```

```python
# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

#### Generate plots

```python
for col in mushroom_data.columns:
    fig, ax = plt.subplots(figsize=(8,6)) 
    ax.set_xlabel(col, size=16)
    ax.set_ylabel('Frequency', size=16)
    ax.set_autoscalex_on(True)
    ax.set_title('{} distribution'.format(col), size=20)
    sns.despine(left=True, bottom=True)
    sns.countplot(mushroom_data[col])
    plt.show()
```
